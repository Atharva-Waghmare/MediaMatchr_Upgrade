```python
import pandas as pd
movies=pd.read_csv('movies_clean_new.csv')
movies
```

```python
from collections import Counter
all_genres=movies['listed_in'].str.split(',').tolist()
all_genres=[genre.strip() for sublist in all_genres for genre in sublist]
all_genres=Counter(all_genres)
all_genres=pd.DataFrame(all_genres.items(), columns=['Genre', 'Count']).sort_values(by='Count', ascending=False)
popular_genres=all_genres.loc[all_genres['Count']>100,'Genre'].tolist()
```

```python
popular_genres
```

```python
movies_with_cat=movies[movies['listed_in'].str.contains('|'.join(popular_genres))]
movies_with_cat[movies_with_cat['Poster_Url']=='Unknown poster link']
```

```python
movies_with_cat.to_csv('movies_with_cat_new.csv', index=False)
```
