```python
import pandas as pd
movies=pd.read_csv('datasets/movies_new.csv',engine='python')
movies.head()
```

```python
movies.rename(columns={'Release_Date':'release_year', 'Vote_Average':'rating', 'Genre':'listed_in','Overview':'description','Title':'title'}, inplace=True)
movies
```

```python
movies['Popularity'].fillna("Unknown popularity", inplace=True)
movies['rating'].fillna("Unknown rating",inplace=True)
movies['Vote_Count'].fillna("Unknown vote count",inplace=True)
movies['Original_Language'].fillna("Unknown language",inplace=True)
movies['listed_in'].fillna("Unknown genre",inplace=True)
movies['Poster_Url'].fillna("Unknown poster link",inplace=True)
movies_with_titles=movies[~movies['title'].isna()]
```

```python
movies_with_titles
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax=plt.axes()
sns.heatmap(movies.isna().transpose())
plt.xlabel("column")
plt.ylabel("row")
plt.show()
```

```python
movies_with_25des=movies_with_titles[movies_with_titles['description'].str.split().str.len()>25]
```

```python
movies_with_25des
```

```python
movies_with_25des.loc[:,'movie_id']=range(1, len(movies_with_25des) + 1)
movies_with_25des.loc[:,'tagged_des']=movies_with_25des[['movie_id','description']].astype(str).agg(' '.join, axis=1)
movies_with_25des
```

```python
movies_with_25des.to_csv('movies_clean_new.csv', index=False)
```
